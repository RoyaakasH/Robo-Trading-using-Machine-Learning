```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('NSE.csv')
x = dataset.iloc[:, 1:4]
dt = dataset [['Date', 'Close']] 
dt.set_index('Date', inplace=True)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, dt_train, dt_test = train_test_split(x, dt, test_size=0.008, random_state=1, shuffle=False )
```

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, dt_train)
```

```
dt_pred = regressor.predict(x_test)
```

```
plt.figure(figsize = (70,20))
plt.plot(dt_test, color =  'green')
plt.plot(dt_pred, color = 'red')
plt.grid(True)
plt.title ('NSE Close Price Representation')
plt.xlabel ('Trading Days')
plt.ylabel ('NSE Close Price')
plt.show ()
```
